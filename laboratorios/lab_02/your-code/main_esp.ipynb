## Ejercicio: Transformación de Palabras

Dada la lista de palabras:

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

1. **Imprimir Letras Mayúsculas**: Escribe un programa que imprima todas las letras de cada palabra en la lista en mayúsculas, separando cada letra por una coma.

```python
# tu pseudocodigo aquí
# dada una lista se recorre cada palabra
# para cada palabra, se recorre con un bucle cada letra y se pasa a mayuscula y se gurada en una nueva lista
# se imprimen las letras la la lista separadas por coma
```

```python
# tu codigo aquí

for palabra in words:
    letras = [letra.upper() for letra in palabra]
    print(", ".join(letras))
```

2. **Crear una Nueva Lista**: Genera una nueva lista que contenga solo las palabras que tengan 5 o más letras. Luego, imprime esta nueva lista.

```python
# tu pseudocodigo aquí
# dada una lista, se recorre con un bucle 
# para cada palabra se comprueba que tenga más de 5 letras
# en tal caso se gurada en una nueva lista
```

```python
# tu codigo aquí
nueva_lista = []
for palabra in words:
    if len(palabra) >= 5: 
        nueva_lista += [palabra]

#otra manera
nueva_lista = [palabra for palabra in words if len(palabra) >= 5]

print(nueva_lista)
```

3. **Encontrar Palabras que Comienzan con 't'**: Imprime la primera palabra en la lista que comience con la letra 't'. Si no hay ninguna, imprime un mensaje indicando que no se encontró ninguna palabra que empiece con 't'.

```python
# tu pseudocodigo aquí
# dada una lista, se recorre con un bucle 
# para cada palabra se comprueba que empiece por t
# se imprime y se incrementa el contador
# si el contador es 0, se imprime que no hay palabras que empiecen por t
```

```python
# tu codigo aquí
count = 0
for palabra in words:
    if palabra.startswith("t"): 
        count+=1
        print(palabra)

if count == 0: print("No hay palabras que empiecen por t")
```

4. **Contar las Vocales**: Escribe un programa que cuente y imprima el número total de vocales en todas las palabras de la lista.

```python
# tu pseudocodigo aquí
# dada una lista, se recorre con un bucle 
# para cada palabra se recorre en bucle las letras
# para cada letra se comprueba si es vocal
# si es vocal se incrementa el contador
# se imprime el contador
```

```python
# tu codigo aquí
vocales = "aeiouAEIOU"
total_vocales = 0

for palabra in words:
    for letra in palabra:
        if letra in vocales:
            total_vocales += 1

print("El número total de vocales es:", total_vocales)
```
