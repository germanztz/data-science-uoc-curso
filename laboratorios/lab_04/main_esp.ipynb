## 8. Escribe una función que devuelva la moda de una lista, es decir: el elemento que aparece más veces.

`NOTE: No se debe usar count... 🧐`

```python
#tu codigo aquí
def mode_imperativa(arr):
    frecuencias = {}
    for elemento in arr:
        if elemento in frecuencias:
            frecuencias[elemento] += 1
        else:
            frecuencias[elemento] = 1
    moda = max(frecuencias, key=frecuencias.get)
    return moda
```

```python
# Para testear tu función
test_mode(mode_imperativa)
```

```python
class Comparador:
    @staticmethod
    def mode_poo(arr):
        frecuencias = {}
        for elemento in arr:
            if elemento in frecuencias:
                frecuencias[elemento] += 1
            else:
                frecuencias[elemento] = 1
        moda = max(frecuencias, key=frecuencias.get)
        return moda

mode_poo = Comparador.mode_poo
```

```python
test_mode(mode_poo)
```

```python
import pandas as pd
mode_funcional = lambda arr: pd.Series(arr).mode()[0]
mode_funcional([1, 2, 2, 3, 3, 3, 4, 4, 5])
```

```python
test_mode(mode_funcional)
```

## 9. Escribe una función que calcule la desviación estándar de una lista.
`NOTE: no utilices librerías ni ninguna función ya construida. 😉`

```python
#tu codigo aquí

def st_dev_imperativa(lista):
    n = len(lista)
    if n == 0:
        return None
    media = sum(lista) / n
    desviacion_cuadrados = sum((x - media) ** 2 for x in lista) / (n - 1)
    return desviacion_cuadrados**(0.5)
```

```python
# Para testear tu función
test_stdev(st_dev_imperativa)
```

```python
class Comparador:
    @staticmethod
    def st_dev_poo(arr):
        n = len(arr)
        if n == 0:
            return None
        media = sum(arr) / n
        desviacion_cuadrados = sum((x - media) ** 2 for x in arr) / (n - 1)
        return desviacion_cuadrados**(0.5)

st_dev_poo = Comparador.st_dev_poo
```

```python
test_stdev(st_dev_poo)
```

```python
st_dev_funcional = lambda arr: (sum((x - sum(arr) / len(arr)) ** 2 for x in arr) / (len(arr) - 1))**(0.5)
```

```python
test_stdev(st_dev_funcional)
```
