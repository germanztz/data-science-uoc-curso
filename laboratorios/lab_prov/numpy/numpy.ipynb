````python
import numpy as np

dates = np.array(['2023-10-01', '2023-10-02', '2023-10-03', '2023-10-04', 
                  '2023-10-05', '2023-10-06', '2023-10-07', '2023-10-08', 
                  '2023-10-09', '2023-10-10', '2023-10-11', '2023-10-12', 
                  '2023-10-13', '2023-10-14', '2023-10-15', '2023-10-16', 
                  '2023-10-17', '2023-10-18', '2023-10-19', '2023-10-20', 
                  '2023-10-21', '2023-10-22', '2023-10-23', '2023-10-24', 
                  '2023-10-25', '2023-10-26', '2023-10-27', '2023-10-28', 
                  '2023-10-29', '2023-10-30', '2023-10-31'])
exchange_rates = np.array([1.10, 1.12, 1.15, 1.11, 1.14, 1.13, 1.10, 1.12, 
                            1.13, 1.11, 1.15, 1.16, 1.18, 1.20, 1.19, 1.17, 
                            1.16, 1.14, 1.12, 1.13, 1.11, 1.10, 1.12, 1.13, 
                            1.14, 1.15, 1.11, 1.10, 1.08, 1.09, 1.11, 1.13])
trade_volumes = np.array([1000, 1200, 1100, 1300, 1500, 1400, 1600, 1700, 
                          1800, 1900, 2100, 2200, 2000, 2300, 2400, 2500, 
                          2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 
                          3400, 3500, 3600, 3700, 3800, 3900, 4000, 4100, 
                          4200])
external_variable = np.array([102, 103, 101, 104, 105, 106, 107, 108, 
                              109, 110, 111, 112, 113, 114, 115, 116, 
                              117, 118, 119, 120, 121, 122, 123, 124, 
                              125, 126, 127, 128, 129, 130, 131, 132, 
                              133])
````

````python
# 1. Calcular la tasa de cambio promedio para todo el mes.
mean_exchange_rate = np.mean(exchange_rates)
print("La tasa de cambio promedio es:", mean_exchange_rate)  # mean_exchange_rate
````

````python
# 2. Identificar la fecha con la tasa de cambio más alta.
max_date = dates[np.argmax(exchange_rates)]
print("La fecha con la tasa de cambio más alta es:", max_date)
````

````python
# 3. Determinar el volumen total de negociación para el mes.
total_trade = np.sum(trade_volumes)
print("El volument total es", total_trade)
````

````python

# 4. Calcular la tasa de cambio mínima y máxima del mes.
min_change_rate = np.min(exchange_rates)
max_change_rate = np.max(exchange_rates)
print("La tasa max es:",max_change_rate, "y la minima es:", min_change_rate)
````

````python

# 5. Determinar la volatilidad diaria de las tasas de cambio (diferencias entre días consecutivos).
volatilidad_diaria = np.diff(exchange_rates)
print(volatilidad_diaria)
````

````python

# 6. Graficar la evolución de las tasas de cambio a lo largo del mes.

import matplotlib.pyplot as plt

dias = np.arange(len(exchange_rates))

plt.figure(figsize=(10, 6))
plt.plot(dias, exchange_rates, marker='o')
plt.title('Evolución de las tasas de cambio')
plt.xlabel('Día')
plt.ylabel('Tasa de cambio')
plt.grid(True)
plt.show()
````

````python

# 7. Comparar el rendimiento de las tasas de cambio en la primera y última semana del mes.
mean_exchange_rate_1 = np.mean(exchange_rates[:7])
mean_exchange_rate_2 = np.mean(exchange_rates[-7:])
print("Diferencia de rendimiento de las tasas de cambio en la primera y última semana del mes es:", mean_exchange_rate_1 - mean_exchange_rate_2)  
````

````python

# 8. Calcular la correlación de las tasas de cambio con respecto a una variable externa (por ejemplo, un índice económico).

# Crear el scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(exchange_rates, external_variable[:-1], color='blue', label='Data Points')

# Etiquetas y título
plt.title('Comparación de la Dispersión: exchange_rates vs external_variable')
plt.xlabel('Exchange Rates')
plt.ylabel('External Variable')

# Mostrar gráfico
plt.legend()
plt.grid(True)
plt.show()
````

````python

# 9. Determinar el número de días en los que la tasa de cambio superó un valor específico (por ejemplo, 1.15).
threshold = 1.15
count = len(list(filter(lambda x: x>threshold,exchange_rates)))
print("días en los que la tasa de cambio superó un valor de", threshold, ":", count )
````

````python

# 10. Crear un resumen estadístico que incluya la media, mediana y desviación estándar de las tasas de cambio.
print(
    "Media:", np.mean(exchange_rates),
    "Mediana:", np.median(exchange_rates),
    "Desviación estándar:", np.std(exchange_rates))
````

````python
# 1. ID del Cliente
# 2. Total de Pedidos Realizados
# 3. Total Gastado (en dólares)
# 4. Días Desde la Última Compra
# 5. Suscrito al Correo Electrónico (1 para suscrito, 0 para no suscrito)

customer_data = np.array([
    [1, 5, 500, 10, 1],
    [2, 3, 300, 20, 1],
    [3, 2, 200, 30, 0],
    [4, 7, 700, 5, 1],
    [5, 4, 400, 15, 1],
    [6, 1, 100, 40, 0],
    [7, 8, 800, 2, 1],
    [8, 6, 600, 25, 0],
    [9, 5, 500, 50, 1],
    [10, 3, 300, 35, 1],
    [11, 0, 0, 60, 0],
    [12, 9, 900, 1, 1],
    [13, 4, 400, 12, 1],
    [14, 2, 200, 18, 0],
    [15, 10, 1000, 0, 1],
    [16, 1, 100, 29, 0],
    [17, 3, 300, 22, 1],
    [18, 5, 500, 14, 1],
    [19, 6, 600, 10, 1],
    [20, 4, 400, 30, 0],
    [21, 7, 700, 4, 1],
    [22, 8, 800, 3, 1],
    [23, 0, 0, 50, 0],
    [24, 2, 200, 45, 1],
    [25, 9, 900, 2, 1],
    [26, 1, 100, 33, 0],
    [27, 6, 600, 7, 1],
    [28, 3, 300, 17, 1],
    [29, 4, 400, 28, 0],
    [30, 5, 500, 19, 1]
])
````

````python
def average_order_value(data):
    '''
    Calcular el valor promedio de pedido (AOV) para todos los clientes. 
    El AOV se calcula como el total gastado dividido por el total de pedidos.
    '''
    total_gastado = np.sum(data[:, 2])
    total_pedidos = np.sum(data[:, 1])
    return np.mean(total_gastado / total_pedidos)

print("AOV:", average_order_value(customer_data))
````

````python
def customer_lifetime_value(data):
    '''
    Calcular el valor de vida útil del cliente (CLV) para cada cliente.
    El CLV se calcula como el producto del valor promedio de pedido y el número de días desde la última compra.
    '''
    return [(cli[2] / cli[1]) * cli[3] if cli[1] > 0 else 0 for cli in data ]

customer_lifetime_value(customer_data)
````

````python
def high_value_customers(data, threshold):
    '''
    Identificar a los clientes de alto valor cuyo CLV excede un umbral dado.
    '''
    high_value_customers = [int(cus[1][0]) for cus in zip(customer_lifetime_value(data), data) if cus[0] > threshold]
    return high_value_customers

high_value_customers(customer_data, 2000)
````

````python
def email_subscription_rate(data):
    '''
    Calcular la tasa de suscripción al correo electrónico, 
    que es el porcentaje de clientes que se han suscrito a los correos electrónicos.
    '''
    return len(list(filter(lambda x: x[4] ==1, data))) / len(data)

email_subscription_rate(customer_data)
````

````python
def inactive_customers(data, days):
    '''
    Identificar a los clientes que no han realizado una compra en un número específico de días.
    '''
    return [int(cus[0]) for cus in data if cus[3] > days]

inactive_customers(customer_data, 30)


````

````python

# ### Actividades Adicionales

# Además de las funciones anteriores, se te pide realizar las siguientes tareas:
def most_frequent_customers(data):
    '''
    Identificar a los clientes que han realizado la mayor cantidad de pedidos. 
    Esto puede ayudar a centrar las estrategias de marketing en los clientes más activos.
    '''
    return data[data[:, 1].argsort()[::-1]]

most_frequent_customers(customer_data)[:5]
    
````

````python
def total_revenue(data):
    '''
    Calcular el ingreso total generado por todos los clientes.
    '''
    return np.sum(data[:, 2])

total_revenue(customer_data)  
````

````python
def average_days_since_last_purchase(data):
    '''
    Calcular el promedio de día desde la última compra para todos los clientes. 
    Esto puede indicar la lealtad y satisfacción del cliente.
    '''
    return np.mean(data[:, 3])

average_days_since_last_purchase(customer_data)

````

````python
def customer_retention_rate(data):
    '''
    Calcular la tasa de retención de clientes,
    que es el porcentaje de clientes que han realizado más de un pedido en comparación con el total de clientes.
    '''
    return len([c for c in data if c[1] > 1]) / len(data)

customer_retention_rate(customer_data)
````

````python
def top_spenders(data, n):
    '''
    Identificar a los `n` principales clientes que han gastado más dinero en total. 
    Esto puede ser útil para estrategias de marketing dirigidas.
    '''
    return data[data[:, 2].argsort()[::-1]][:n]

top_spenders(customer_data, 10)
````
